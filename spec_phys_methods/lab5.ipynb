```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sym = pd.read_csv('./Simetric.TXT', sep='\t', skiprows=19, names=['theta', 'Симетрична'])
asym = pd.read_csv('./Asimetric.TXT', sep='\t', skiprows=19, names=['theta', 'Асиметрична'])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
asym.plot(x='theta', y='Асиметрична', ax=ax, linestyle='-')
sym.plot(x='theta', y='Симетрична', ax=ax, linestyle='--')
plt.ylim((0, 120))
plt.xlim((35, 48))
plt.xlabel(r'Кут дифракції $2\theta, {}^{\circ}$')
plt.ylabel(r'Інтенсивність, в.о.')
plt.grid()
plt.tight_layout()
fig.savefig('lab5.pdf')
plt.show()
```
