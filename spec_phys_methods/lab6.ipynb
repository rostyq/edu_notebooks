```python
kowz = pd.read_csv('./data/lab6/Kowz.TXT', sep='\t', skiprows=19, names=['theta', 'Ковзаючий пучок'])
bb = pd.read_csv('./data/lab6/Lab6.TXT', sep='\t', skiprows=19, names=['theta', 'Брегга-Брентано'])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 5))
kowz.plot(x='theta', y='Ковзаючий пучок', ax=ax, linestyle='-')
bb.plot(x='theta', y='Брегга-Брентано', ax=ax, linestyle='--')
plt.ylim((0, 170))
plt.xlim((35, 53))
plt.xlabel(r'$\theta, {}^{\circ}$')
plt.ylabel(r'$Інтенсивність, в.о.$')
plt.grid()
plt.tight_layout()
fig.savefig('lab6.pdf')
plt.show()
```
